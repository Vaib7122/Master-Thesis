```python
import os
import zipfile
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img

# Set up directories
def extract_zip(zip_path, extract_to):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)

# Preprocessing function for images
def preprocess_image(image_path, target_size=(224, 224)):
    image = load_img(image_path, target_size=target_size)
    image = img_to_array(image)
    image = image / 255.0
    return image

# Load and preprocess images
def load_and_preprocess_images(dataset_dir, target_size=(224, 224)):
    images = []
    labels = []
    class_names = []
    for class_name in os.listdir(dataset_dir):
        class_path = os.path.join(dataset_dir, class_name)
        if os.path.isdir(class_path):
            class_names.append(class_name)
            for img_name in os.listdir(class_path):
                img_path = os.path.join(class_path, img_name)
                try:
                    image = preprocess_image(img_path, target_size)
                    images.append(image)
                    labels.append(class_name)
                except Exception as e:
                    print(f"Error processing image {img_path}: {e}")
    return np.array(images), labels, class_names

# K-Means segmentation
# def segment_images_with_kmeans(images, n_clusters=3):
#     segmented_images = []
#     for img in images:
#         pixels = img.reshape(-1, 3)
#         kmeans = KMeans(n_clusters=n_clusters, random_state=42).fit(pixels)
#         segmented_pixels = kmeans.cluster_centers_[kmeans.labels_]
#         segmented_image = segmented_pixels.reshape(img.shape)
#         return segmented_image

    #     segmented_images.append(segmented_image)
    # return np.array(segmented_images)



# K-Means segmentation for a single image
def segment_image_with_kmeans(image, n_clusters=3):
    pixels = image.reshape(-1, 3)
    kmeans = KMeans(n_clusters=n_clusters, random_state=42).fit(pixels)
    segmented_pixels = kmeans.cluster_centers_[kmeans.labels_]
    segmented_image = segmented_pixels.reshape(image.shape)
    return segmented_image


# Build the DenseNet model
def build_densenet_model(num_classes):
    base_model = DenseNet121(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(512, activation='relu')(x)
    predictions = Dense(num_classes, activation='softmax')(x)
    model = Model(inputs=base_model.input, outputs=predictions)
    for layer in base_model.layers:
        layer.trainable = False
    model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
    return model


# Prediction function for a new image
def predict_disease(image_path, model, label_encoder):
    # Preprocess and segment the image
    image = preprocess_image(image_path)
    segmented_image = segment_image_with_kmeans(image)
    segmented_image = np.expand_dims(segmented_image, axis=0)  # Add batch dimension

    # Predict using the model
    predictions = model.predict(segmented_image)
    predicted_index = np.argmax(predictions)
    predicted_label = label_encoder.inverse_transform([predicted_index])

    # Display results
    print(f"Raw Predictions: {predictions}")
    print(f"Predicted Class Index: {predicted_index}")
    print(f"Predicted Disease: {predicted_label[0]}")


# Main pipeline
def main():
    # Paths
    zip_path = "/content/drive/MyDrive/coco diseas.zip"  # Path to your ZIP file
    extract_to = "dataa"

    # Extract ZIP file
    extract_zip(zip_path, extract_to)
    dataset_dir = extract_to

    # Load and preprocess images
    images, labels, class_names = load_and_preprocess_images(dataset_dir)
    if len(images) == 0 or len(labels) == 0:
        print("No images or labels found!")
        return

    print(f"Loaded {len(images)} images with {len(set(labels))} classes.")

    # Segment images using K-Means
    # segmented_images = segment_images_with_kmeans(images)
    print("Segmented images using K-Means.")

    # Encode labels
    label_encoder = LabelEncoder()
    encoded_labels = label_encoder.fit_transform(labels)
    categorical_labels = to_categorical(encoded_labels)

    # Split data into train and validation
    from sklearn.model_selection import train_test_split
    X_train, X_val, y_train, y_val = train_test_split(images, categorical_labels, test_size=0.2, random_state=42)

    # Build and train DenseNet model
    cnn_model = build_densenet_model(len(class_names))
    history = cnn_model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, batch_size=16)

    # Evaluate model
    test_loss, test_accuracy = cnn_model.evaluate(X_val, y_val)
    print(f"Test Accuracy: {test_accuracy * 100:.2f}%")


     # Save the model and label encoder
    cnn_model.save("disease_model.h5")
    np.save("label_encoder_classes.npy", label_encoder.classes_)

    # Predict disease from a new image
    sample_image_path = "/content/dataa/stem Bleeding/StemBleeding183.jpg"  # Update with the path to your test image
    predict_disease(sample_image_path, cnn_model, label_encoder)


    # sample_image_path = "/content/data/stem Bleeding/StemBleeding172.jpg"
    # preprocessed_image = preprocess_image(sample_image_path)

    # Predict using the CNN model
    # predictions =  cnn_model
    # predicted_index = np.argmax(predictions)
    # predicted_disease = label_encoder.inverse_transform([predicted_index])

    # print("Raw Predictions:", predictions)
    # print("Predicted Index:", predicted_index)
    # print("Predicted Disease:", predicted_disease)




    # Predictions
    y_pred = cnn_model.predict(X_val)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true_classes = np.argmax(y_val, axis=1)

    # Classification report
    print(classification_report(y_true_classes, y_pred_classes, target_names=class_names))

    # Confusion Matrix
    cm = confusion_matrix(y_true_classes, y_pred_classes)
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion Matrix')
    plt.colorbar()
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import os
import zipfile
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
from sklearn.metrics import roc_curve, auc, roc_auc_score
from sklearn.preprocessing import label_binarize

# Set up directories
def extract_zip(zip_path, extract_to):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)

# Preprocessing function for images
def preprocess_image(image_path, target_size=(224, 224)):
    image = load_img(image_path, target_size=target_size)
    image = img_to_array(image)
    image = image / 255.0
    return image

# Load and preprocess images
def load_and_preprocess_images(dataset_dir, target_size=(224, 224)):
    images = []
    labels = []
    class_names = []
    for class_name in os.listdir(dataset_dir):
        class_path = os.path.join(dataset_dir, class_name)
        if os.path.isdir(class_path):
            class_names.append(class_name)
            for img_name in os.listdir(class_path):
                img_path = os.path.join(class_path, img_name)
                try:
                    image = preprocess_image(img_path, target_size)
                    images.append(image)
                    labels.append(class_name)
                except Exception as e:
                    print(f"Error processing image {img_path}: {e}")
    return np.array(images), labels, class_names

# K-Means segmentation for a single image
def segment_image_with_kmeans(image, n_clusters=3):
    pixels = image.reshape(-1, 3)
    kmeans = KMeans(n_clusters=n_clusters, random_state=42).fit(pixels)
    segmented_pixels = kmeans.cluster_centers_[kmeans.labels_]
    segmented_image = segmented_pixels.reshape(image.shape)
    return segmented_image


# Build the DenseNet model
def build_densenet_model(num_classes):
    base_model = DenseNet121(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(512, activation='relu')(x)
    predictions = Dense(num_classes, activation='softmax')(x)
    model = Model(inputs=base_model.input, outputs=predictions)
    for layer in base_model.layers:
        layer.trainable = False
    model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
    return model


# Prediction function for a new image
def predict_disease(image_path, model, label_encoder):
    # Preprocess and segment the image
    image = preprocess_image(image_path)
    segmented_image = segment_image_with_kmeans(image)
    segmented_image = np.expand_dims(segmented_image, axis=0)  # Add batch dimension

    # Predict using the model
    predictions = model.predict(segmented_image)
    predicted_index = np.argmax(predictions)
    predicted_label = label_encoder.inverse_transform([predicted_index])

    # Display results
    print(f"Raw Predictions: {predictions}")
    print(f"Predicted Class Index: {predicted_index}")
    print(f"Predicted Disease: {predicted_label[0]}")


# Main pipeline
def main():
    # Paths
    zip_path = "/content/drive/MyDrive/coco diseas.zip"  # Path to your ZIP file
    extract_to = "dataa"

    # Extract ZIP file
    extract_zip(zip_path, extract_to)
    dataset_dir = extract_to

    # Load and preprocess images
    images, labels, class_names = load_and_preprocess_images(dataset_dir)
    if len(images) == 0 or len(labels) == 0:
        print("No images or labels found!")
        return

    print(f"Loaded {len(images)} images with {len(set(labels))} classes.")

    # Encode labels
    label_encoder = LabelEncoder()
    encoded_labels = label_encoder.fit_transform(labels)
    categorical_labels = to_categorical(encoded_labels)

    # Split data into train and validation
    from sklearn.model_selection import train_test_split
    X_train, X_val, y_train, y_val = train_test_split(images, categorical_labels, test_size=0.2, random_state=42)

    # Build and train DenseNet model
    cnn_model = build_densenet_model(len(class_names))
    history = cnn_model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=10, batch_size=16)

    # Evaluate model
    test_loss, test_accuracy = cnn_model.evaluate(X_val, y_val)
    print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

    # Save the model and label encoder
    cnn_model.save("disease_model.h5")
    np.save("label_encoder_classes.npy", label_encoder.classes_)

    # Ensure the directory exists before saving
    model_dir = "/content/saved_models"  # Choose a directory to save your model
    os.makedirs(model_dir, exist_ok=True)  # Create the directory if it doesn't exist
    model_path = os.path.join(model_dir, "disease_model.h5")
    label_encoder_path = os.path.join(model_dir, "label_encoder_classes.npy")

    cnn_model.save(model_path)
    np.save(label_encoder_path, label_encoder.classes_)


    # Predict disease from a new image
    sample_image_path = "/content/dataa/stem Bleeding/StemBleeding183.jpg"  # Update with the path to your test image
    predict_disease(sample_image_path, cnn_model, label_encoder)

    # Predictions
    y_pred = cnn_model.predict(X_val)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true_classes = np.argmax(y_val, axis=1)

    # Classification report
    print(classification_report(y_true_classes, y_pred_classes, target_names=class_names))

    # Confusion Matrix
    cm = confusion_matrix(y_true_classes, y_pred_classes)
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion Matrix')
    plt.colorbar()
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()

    # --- ROC Curve Calculation ---
    y_true_bin = y_val  # This is already one-hot encoded in your case
    n_classes = len(class_names)

    # Compute ROC curve and ROC area for each class (moved outside loop for clarity)
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

 # Check if y_true_bin and y_pred have the correct shape
    print("Shape of y_true_bin:", y_true_bin.shape)
    print("Shape of y_pred:", y_pred.shape)

 # Iterate through classes and calculate ROC curve for each
    for i in range(n_classes):
        # Check if the class has enough samples for ROC calculation
        num_samples_in_class = np.sum(y_true_bin[:, i])
        if num_samples_in_class > 0:  # Only calculate ROC if class has samples
            fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_pred[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])
        else:
            print(f"Skipping ROC calculation for class {i} due to insufficient samples.")
            fpr[i], tpr[i], roc_auc[i] = [], [], np.nan  # Set empty values for missing classes


    # Calculate macro-average ROC curve and ROC area
    # Get list of classes for which we have ROC curves
    valid_classes = [i for i in range(n_classes) if len(fpr[i]) > 0]


      # Calculate macro-average ROC only for valid classes
    if valid_classes: # Only calculate macro-average if valid classes exist
        all_fpr = np.unique(np.concatenate([fpr[i] for i in valid_classes]))
        mean_tpr = np.zeros_like(all_fpr)
        for i in valid_classes:
            mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
        mean_tpr /= len(valid_classes)
        fpr["macro"] = all_fpr
        tpr["macro"] = mean_tpr
        roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])
    else:
        print("No valid classes for macro-average ROC calculation.")


    # all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
    # mean_tpr = np.zeros_like(all_fpr)
    # for i in range(n_classes):
    #     mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
    # mean_tpr /= n_classes
    # fpr["macro"] = all_fpr
    # tpr["macro"] = mean_tpr
    # roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

    # # Plot ROC curves
    # plt.figure(figsize=(10, 8))
    # plt.plot(
    #     fpr["micro"],
    #     tpr["micro"],
    #     label="micro-average ROC curve (area = {0:0.2f})".format(roc_auc["micro"]),
    #     color="deeppink",
    #     linestyle=":",
    #     linewidth=4,
    # )

    # plt.plot(
    #     fpr["macro"],
    #     tpr["macro"],
    #     label="macro-average ROC curve (area = {0:0.2f})".format(roc_auc["macro"]),
    #     color="navy",
    #     linestyle=":",
    #     linewidth=4,
    # )

    # colors = cycle(["aqua", "darkorange", "cornflowerblue"])
    # for i, color in zip(range(n_classes), colors):
    #     plt.plot(
    #         fpr[i],
    #         tpr[i],
    #         color=color,
    #         lw=lw,
    #         label="ROC curve of class {0} (area = {1:0.2f})".format(i, roc_auc[i]),
    #     )

    # plt.plot([0, 1], [0, 1], "k--", lw=lw)
    # plt.xlim([0.0, 1.0])
    # plt.ylim([0.0, 1.05])
    # plt.xlabel("False Positive Rate")
    # plt.ylabel("True Positive Rate")
    # plt.title("Receiver operating characteristic to multi-class")
    # plt.legend(loc="lower right")
    # plt.show()

    return history, cnn_model, X_val, label_encoder, y_true_classes, y_pred # Return necessary components

if __name__ == "__main__":
    history, cnn_model, X_val, label_encoder, y_true_classes, y_pred = main()

    # Now you can plot the accuracy
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()


    # --- Calculate ROC AUC ---
    y_pred_prob = cnn_model.predict(X_val)  # Get predicted probabilities
    # y_true_classes = np.argmax(y_val, axis=1) # Get true labels (This is already defined in 'main')
    # n_classes = len(label_encoder.classes_) # (This is already defined in 'main')
    class_names = label_encoder.classes_
    y_true_bin = label_binarize(y_true_classes, classes=np.arange(len(class_names)))
    # y_pred_prob = label_binarize(y_pred_prob, classes=np.arange(len(class_names)))
    # Compute ROC curve and ROC area for each class
    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(len(class_names)):
        fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_pred_prob[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # Compute micro-average ROC curve and ROC area
    fpr["micro"], tpr["micro"], _ = roc_curve(y_true_bin.ravel(), y_pred_prob.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

    # # Plot ROC curves for each class or micro-average
    # plt.figure()
    # # ... (Plotting code for ROC curves) ...
    # plt.show()


     # Plot ROC curves for each class or micro-average
    plt.figure(figsize=(10, 8))  # Adjust figure size if needed

    # Plot micro-average ROC curve
    plt.plot(fpr["micro"], tpr["micro"],
             label='micro-average ROC curve (area = {0:0.2f})'
                   ''.format(roc_auc["micro"]),
             color='deeppink', linestyle=':', linewidth=4)

    # Plot ROC curves for individual classes
    for i in range(len(class_names)):
        plt.plot(fpr[i], tpr[i], lw=2,
                 label='ROC curve of class {0} (area = {1:0.2f})'
                       ''.format(class_names[i], roc_auc[i]))

    plt.plot([0, 1], [0, 1], 'k--', lw=2)  # Diagonal line
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic to multi-class')
    plt.legend(loc="lower right")
    plt.show()


```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import joblib

# Load model and label encoder using the correct filenames and paths
model_dir = "/content/saved_models" # The directory where you saved the model
model_path = os.path.join(model_dir, "disease_model.h5")
label_encoder_path = os.path.join(model_dir, "label_encoder_classes.npy")

model = load_model(model_path)
label_encoder = LabelEncoder()  # Initialize a LabelEncoder object
label_encoder.classes_ = np.load(label_encoder_path, allow_pickle=True) # Set the classes attribute


# Get class names
class_names = list(label_encoder.classes_)
```

```python
from tensorflow.keras.preprocessing import image
from sklearn.cluster import KMeans
import cv2

def preprocess_and_segment(img_path):
    img = cv2.imread(img_path)
    img = cv2.resize(img, (224, 224))
    # Apply K-Means
    img_reshaped = img.reshape((-1, 3))
    kmeans = KMeans(n_clusters=3, random_state=42).fit(img_reshaped)
    segmented_img = kmeans.cluster_centers_[kmeans.labels_].reshape(img.shape).astype(np.uint8)
    # Normalize and prepare for prediction
    img_array = segmented_img / 255.0
    return np.expand_dims(img_array, axis=0)  # Add batch dimension

# Example
sample_image = preprocess_and_segment("/content/dataa/Leaf rot/LeafRot001.jpg")
```

```python
# Predict
pred_probs = model.predict(sample_image)[0]  # Single image, so index 0
predicted_class = class_names[np.argmax(pred_probs)]

print(f"Predicted Class: {predicted_class}")
print(f"Confidence Scores: {pred_probs}")
```

```python
def plot_softmax_bar(probs, class_names):
    plt.figure(figsize=(8, 5))
    bars = plt.bar(class_names, probs, color='skyblue')
    plt.title("Softmax Probabilities for Each Class")

    plt.ylabel("Confidence")
    plt.ylim([0, 1])
    plt.xticks(rotation=45)

    # Annotate the bars
    for bar, prob in zip(bars, probs):
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.01, f'{prob:.2f}', ha='center', va='bottom')

    plt.tight_layout()
    plt.show()

plot_softmax_bar(pred_probs, class_names)
```

```python
def plot_multiple_softmax(predictions_list, class_names, image_labels=None):
    """
    predictions_list: List of softmax output arrays (e.g., from multiple test images)
    image_labels: Optional list of image names for x-axis labels
    """
    num_images = len(predictions_list)
    num_classes = len(class_names)
    x = np.arange(num_classes)

    fig, axs = plt.subplots(num_images, 1, figsize=(8, 4 * num_images))

    if num_images == 1:
        axs = [axs]  # wrap in list if single

    for i, (probs, ax) in enumerate(zip(predictions_list, axs)):
        ax.bar(class_names, probs, color='lightcoral')
        ax.set_ylim([0, 1])
        title = f"Image {i+1}" if not image_labels else f"{image_labels[i]}"
        ax.set_title(f"Softmax Probabilities  {title}")
        for j, p in enumerate(probs):
            ax.text(j, p + 0.01, f"{p:.2f}", ha='center')

    plt.tight_layout()
    plt.show()
```
